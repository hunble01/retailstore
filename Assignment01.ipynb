```python
#Lab Professor: Reza
employee_list = []
item_list = []


# Function to create employees
def create_employee():
    while True:
        employee_id = input("Enter Employee ID (or 'NO' to finish): ")
        if employee_id.upper() == 'NO':
            break

        # Validate employee_id is unique and numeric
        if not employee_id.isdigit() or int(employee_id) in [employee[0] for employee in employee_list]:
            print("Invalid employee ID. Please enter a unique numeric value.")
            continue

        employee_name = input("Enter Employee Name: ")
        employee_type = input("Enter Employee Type (hourly/manager): ").lower()

        # Validate employee_type
        if employee_type not in ['hourly', 'manager']:
            print("Invalid employee type. Please enter 'hourly' or 'manager'.")
            continue

        years_worked = input("Enter Years Worked: ")

        # Validate years_worked is numeric
        if not years_worked.isdigit():
            print("Invalid years worked. Please enter a numeric value.")
            continue

        employee_discount_number = input("Enter Employee Discount Number: ")

        # Validate employee_discount_number is unique and numeric
        if not employee_discount_number.isdigit() or int(employee_discount_number) in [employee[6] for employee in employee_list]:
            print("Invalid employee discount number. Please enter a unique numeric value.")
            continue

        # Add employee to the list
        employee_list.append([int(employee_id), employee_name, employee_type, int(years_worked), 0, 0, int(employee_discount_number)])

        add_another = input("Do you want to add another employee? (yes/no): ")
        if add_another.lower() != 'yes':
            break

# Function to update employee information based on purchases
def update_employee_info(employee_discount_number, total_purchased, total_discount):
    for employee in employee_list:
        if employee[6] == employee_discount_number:
            employee[4] += total_purchased
            employee[5] += total_discount
            break

# Function to create items
def create_item():
    while True:
        item_number = input("Enter Item Number (or 'NO' to finish): ")
        if item_number.upper() == 'NO':
            break

        # Validate item_number is unique and numeric
        if not item_number.isdigit() or int(item_number) in [item[0] for item in item_list]:
            print("Invalid item number. Please enter a unique numeric value.")
            continue

        item_name = input("Enter Item Name: ")
        item_cost = input("Enter Item Cost: ")

        # Validate item_cost is numeric
        if not item_cost.replace('.', '', 1).isdigit():
            print("Invalid item cost. Please enter a numeric value.")
            continue

        # Add item to the list
        item_list.append([int(item_number), item_name, float(item_cost)])

        add_another = input("Do you want to add another item? (yes/no): ")
        if add_another.lower() != 'yes':
            break

# Function to make purchases
def make_purchase():
    while True:
        # Display available items
        print("Item Number | Item Name         | Item Cost")
        for item in item_list:
            print(f"{item[0]}{' ' * (13 - len(str(item[0])))}| {item[1]}{' ' * (17 - len(item[1]))}| ${item[2]:.2f}")

        item_number = input("Enter Item Number to purchase (or 'NO' to finish): ")
        if item_number.upper() == 'NO':
            break

        # Validate item_number is valid and numeric
        if not item_number.isdigit() or int(item_number) not in [item[0] for item in item_list]:
            print("Invalid item number. Please enter a valid numeric value.")
            continue

        quantity = input("Enter Quantity: ")

        # Validate quantity is numeric and positive
        if not quantity.isdigit() or int(quantity) <= 0:
            print("Invalid quantity. Please enter a positive numeric value.")
            continue

        employee_discount_number = input("Enter Employee Discount Number: ")

        # Validate employee_discount_number is valid and numeric
        if not employee_discount_number.isdigit() or int(employee_discount_number) not in [employee[6] for employee in employee_list]:
            print("Invalid employee discount number. Please enter a valid numeric value.")
            continue

        # Find item and calculate total price
        selected_item = next(item for item in item_list if item[0] == int(item_number))
        base_price = float(selected_item[2]) * int(quantity)

        # Apply employee discount
        for employee in employee_list:
            if employee[6] == int(employee_discount_number):
                discount_rate = min(employee[3] * 0.02, 0.10)  # 2% for each year, maximum 10%
                if employee[2] == 'manager':
                    discount_rate += 0.10  # Additional 10% discount for managers
                elif employee[2] == 'hourly':
                    discount_rate += 0.02  # 2% discount for hourly employees

                total_price = base_price - (base_price * discount_rate)

                # Apply maximum discount limit of $200
                total_price = max(total_price, 0) if employee[5] + total_price > 200 else total_price

                print(f"Total Price: ${total_price:.2f}")

                # Update employee information based on purchase
                update_employee_info(int(employee_discount_number), base_price, base_price - total_price)
                break

        make_another = input("Do you want to make another purchase? (yes/no): ")
        if make_another.lower() != 'yes':
            break

# Function to display all-employee summary
def all_employee_summary():
    print("Employee ID | Employee Name | Employee Type | Years Worked | Total Purchased | Total Discount | Employee Discount Number")
    for employee in employee_list:
        print(f"{employee[0]}{' ' * (12 - len(str(employee[0])))}| {employee[1]}{' ' * (16 - len(employee[1]))}| {employee[2]}{' ' * (14 - len(employee[2]))}| {employee[3]}{' ' * (13 - len(str(employee[3])))}| ${employee[4]:.2f}{' ' * (16 - len(str(employee[4])))}| ${employee[5]:.2f}{' ' * (15 - len(str(employee[5])))}| {employee[6]}")

# Main program loop
while True:
    print("\n-------------------------------")
    print("|        Retail System        |")
    print("-------------------------------")
    print("|   1- Create Employee        |")
    print("|   2- Create Item            |")
    print("|   3- Make Purchase          |")
    print("|   4- All Employee Summary   |")
    print("|   5- Exit                   |")
    print("-------------------------------")

    choice = input("Enter your choice: ")

    if choice == '1':
        create_employee()
    elif choice == '2':
        create_item()
    elif choice == '3':
        make_purchase()
    elif choice == '4':
        all_employee_summary()
    elif choice == '5':
        print("Exiting the program. Goodbye!")
        break
    else:
        print("Invalid choice. Please enter a number between 1 and 5.")






```
